```python
import scipy.io as sio 
import numpy as np
from keras.models import Graph, Sequential
from keras.layers.core import Dense, Dropout, Activation, TimeDistributedDense
from keras.layers.embeddings import Embedding
from keras.layers.recurrent import GRU, LSTM
from keras.layers.noise import GaussianNoise, GaussianDropout
from keras.callbacks import ModelCheckpoint
from keras.regularizers import l2
import re
import csv
import os
from liblinearutil import *
from tsne import *
%matplotlib inline
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
def plt_tsne(bag):
    tsne_cor=bag['tsne_cor']
    l=bag['train_len']
    pre_idx=bag['pre_idx']
    pre_tsneY=bag['pre_tsneY']
    plt.scatter(tsne_cor[:l,0], tsne_cor[:l,1], s=30, c=pre_tsneY[:l], marker='>', alpha=0.7)
    plt.scatter(tsne_cor[l:,0], tsne_cor[l:,1], s=30, c=pre_tsneY[l:], marker='o', alpha=0.7)
    edge_person=np.asarray([34,116,368,339,65,366,58,120,158,86])
    for edge_one in edge_person:
        edge_index=np.where(pre_idx==edge_one)
        #print edge_index[0], l
        if edge_index[0]>l:
            plt.scatter(tsne_cor[edge_index, 0], tsne_cor[edge_index, 1], s=40, facecolors='none', edgecolors='y')
    plt.show()
```

```python
his_tsne=np.load('tsne_07result.npy')
for bag in his_tsne:
    plt_tsne(bag)
```
